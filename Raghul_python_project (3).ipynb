#3) Exploratory Data Analysis (EDA) and Visualization
##Campaign Performance:

```python
plt.figure(figsize=(17,6))
plt.scatter(pd1['Acquisition_Cost'],pd1['ROI'],linewidths=2.8)
plt.title('Acquisition_Cost and ROI')
plt.xlabel('Acquisition_Cost')
plt.ylabel('ROI')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(17,6))
sns.barplot(data=pd1,x='Conversion_Rate',y='Channel_Used',hue='Campaign_Type')
plt.title('average Conversion_Rate in Channel_Used of Campaign_Type')
plt.grid()
plt.xlabel('Conversion_Rate')
plt.ylabel('Channel_Used')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(data=pd1,x='Engagement_Score',y='Campaign_Type')
plt.xlabel('Engagement_Score')
plt.ylabel('Campaign_Type')
plt.title('Engagement_Score across different Campaign_Type')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(17,5))
average_ROI = pd1.groupby('Company')['ROI'].mean()
average_ROI.plot(kind='bar')
plt.xlabel('Company')
plt.ylabel('ROI')
plt.grid()
plt.title('ROI by Company using')
plt.show()
```

```python
plt.figure(figsize=(15,5))
correction = pd1[['Engagement_Score','Conversion_Rate']].corr()
sns.heatmap(correction,annot=True,cmap='coolwarm',vmin=1,vmax=1)
plt.title('Engagement_Score correction Conversion_Rate')
plt.grid()
plt.show()
```

#Customer Segmentation:

```python
plt.figure(figsize=(8,11))
sns.countplot(data=pd1,x='Target_Audience')
plt.xticks(rotation=45)
plt.title('the distribution of Target_Audience')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.barplot(data=pd1,x='Language',y='Conversion_Rate',estimator='mean',hue='Customer_Segment')
plt.title('conversion rate by customer segment and language')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.boxplot(data=pd1,x='Customer_Segment',y='Acquisition_Cost',hue='Channel_Used')
plt.title('acquisition cost by customer segment and channel used')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.barplot(data=pd1,x='Language',y='Conversion_Rate',estimator='mean')
plt.title('average conversion rate by language')
plt.grid()
plt.show()
```

#Channel Effectiveness:

```python
plt.figure(figsize=(15,6))
sns.barplot(data=pd1,x='Channel_Used',y='Engagement_Score',hue='Campaign_Type')
plt.title('engagement score diffrent channel user by campaign type ')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,10))
total_ROI = pd1.groupby('Channel_Used')['ROI'].sum()
total_ROI.plot(kind='pie',autopct='%1.3f%%',startangle=90)
plt.title('Total ROI diffrent channel used ')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(17,7))
sns.scatterplot(data=pd1,x='Clicks',y='Impressions',hue='Channel_Used')
plt.title(' the relationship between Clicks and Impressionsa each Channel_Used')
plt.grid()
plt.tight_layout()
plt.show()
```

#Time-Based Analysis:

```python
plt.figure(figsize=(15,7))
sns.histplot(data=pd1,x='Duration',color= 'y')
plt.title('the distribution of Duration')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(data=pd1,x='Date',y='Conversion_Rate',hue='Company')
plt.title('the overall Conversion Rate over Date each company')
plt.xticks('DATE')
plt.grid()
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(20,6))
sns.lineplot(data=pd1,x='Date',y='Engagement_Score',color='k')
plt.title('the trend of Engagement Score over Date')
plt.tight_layout()
plt.show()
```

#Geographic Analysis:

```python
plt.figure(figsize=(16,6))
sns.barplot(data=pd1,x='Location',y='Acquisition_Cost')
plt.title('list of  location has the highest Acquisition_Cost')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.barplot(data=pd1,x='Conversion_Rate',y='Location',hue='Target_Audience')
plt.title('Conversion Rate  different Location category Target Audience ')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,10))
Sum_ROI = pd1.groupby('Location')['ROI'].sum()
Sum_ROI.plot(kind='pie',autopct='%1.3f%%')
plt.title('Illustrate the proportion of ROI by Location')
plt.grid()
plt.show()
```
